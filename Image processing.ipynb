# **Manip 2:**

1- creation de l'image couleur:

```python
#l'image couleur contient 7 couleurs, et chaque couleur est representer sur les 3 axes des couleurs de bases rouge vert et bleu
# l'image est de dimention 28x28 alors chaque couleur vas prendre 4 pixels suivant les colonnes et 28 pixels en lignes
# on doit cree trois vecteurs ou chaque vecteur represente un axes R ou G ou B 
# Chaque axe est un vecteur de 28 elements et chaque element represente la valeur du pixel paraport a cette couleur.

A=np.array([255,255,255,255,0,0,0,0,0,0,0,0,255,255,255,255,255,255,255,255,0,0,0,0,255,255,255,255])
B=np.array([0,0,0,0,255,255,255,255,0,0,0,0,255,255,255,255,255,255,255,255,255,255,255,255,0,0,0,0])
C=np.array([0,0,0,0,0,0,0,0,255,255,255,255,255,255,255,255,0,0,0,0,255,255,255,255,255,255,255,255])
R=np.tile(A,(28,1))#le vecteur R est le rouge 
G=np.tile(B,(28,1)) #le vecteur G est le rouge
B=np.tile(C,(28,1))#le vecteur B est le rouge
print(type(A[0]))


#apres on doit fusionner les trois vecteurs de couleurs pour obtenir la matrice de l'image couleur
I=cv.merge([R,G,B])
np.uint8(I)
plt.imshow(I)#affichage de l'image couleur
plt.show()
I.size


```

1 - Chaque pixel de l'image obtenue est codee sur 24 bits (8 bits pour chauqe plan RGB)

2 - Le poids théorique de l'image est : 28x28x3=**2352 octets**

3- Enregistrement de l'image :

```python
cv.imwrite("image_tp_bmp.bmp",I)
```



*   Le poids du fichier enregistrer est :**2406 Octets**

4 - Séparation les canaux RGB :

```python
R,G,B=cv.split(img)
#affichage des image separee:
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(R,cmap='gray')
ax1.set_title('plan Rouge')
ax2.imshow(G,cmap='gray')
ax2.set_title('plan vert')
ax3.imshow(B,cmap='gray')
ax3.set_title('plan bleu')
plt.tight_layout()
plt.show()

```

**Commentaire:**


*   Les yeux de maderill dans le plan rouge apparaissent en gris clair, on pla vert en gris et en plan bleu il sont presque noirs.
*   Cela est expliquée par le fait que ses yeux sont de couleurs orange qui est une combinaison entre le vert et le rouge avec une intensite plus elevée du rouge et c'est ce qu'on voit sur les trois images la valeur du rouge est elevée, c'est pour ça elle apparait en gris clair, la valeur du vert est moyenne c'est pour ça elle est en gris et la valeur du bleu ce qui justefie pourquoi elle est en noir.




```python


figure, ax = plt.subplots(2,3)
img1=cv.merge([R,G,B])
ax[0,0].imshow(img1)

img2=cv.merge([G,R,B])
ax[0,1].imshow(img2)

img3=cv.merge([B,G,R])
ax[0,2].imshow(img3)

img4=cv.merge([B,R,G])
ax[1,0].imshow(img4)

img5=cv.merge([G,B,R])
ax[1,1].imshow(img5)

img6=cv.merge([R,B,G])
ax[1,2].imshow(img6)
plt.tight_layout()
plt.show()

```

Commentaire:
 


*   On voit bien que les couleurs sont differentes dans
chacune des images obtenues, et cela est due que chaque matrice R,ou G ,B contient des nuance de chaque couleur respectivement, alors lorsqu'on change l'emblacement de ces matrices dans la fonction merge elle change automatiquement les valeurs de rgb pour chaque pixel et donc les couleurs de l'image.
